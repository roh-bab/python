#Data Visualisation refers to the graphical or pictorial representstion of data using graphs,charts,maps etc.The purpose of visualisation is to visualise variations or show relationship between variables making data more accessible, understandable and easy to use.

#Matplotlib is a powerfull and versatile plotting library for python that enables us to create static , animated and interactive visualisations.Matplotlib interacts with numpy arrays and pandas DataFrames making it convinient to visualise data in this format.


```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
print('Creating a plot')
x = [10,20,30,40,50,60]
y = [15,30,45,60,75,90]
print('x=', x)
print('y=',y)
plt.plot(x,y,label = 'first plot')
plt.title('Simple line plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

print('Show the plot')
plt.show()
```

```python
weight = [19.7, 21.3, 23.5, 25.9, 28.5, 32.1, 35.7, 39.6, 43.2]
height = [121.9, 124.5, 129.5, 134.6, 139.7, 147.3, 152.4, 157.5,162.6]
#plot setting
plt.figure(figsize=[8,6])
plt.plot(weight,height,color= 'blue',marker='*',markersize= 6,linestyle='--',linewidth= 1,label = 'Weight vs height')
plt.xlabel= 'weight'
plt.ylabel= 'height'
plt.grid(True)
plt.show()
```

```python
print('Using matplotlib with numpy array')
import numpy as np
x = np.array([1,2,3,4,5,6])
y = x ** 2
plt.plot(x,y,label= 'XY',color= 'green',marker= 'o')
#plt.xlabel('X axis')
#plt.ylabel('Y axis')
plt.grid(True)
plt.xticks(np.arange(0,6,step=1))
plt.yticks(np.arange(0,41,step=5))
plt.show()
```

#Matplotlib with DataFrame, We must make sure that there are no null values in the dataset.

```python
from csv import Error
import pandas as pd
try:
  customers = pd.read_excel('customers.xlsx')
  products =  pd.read_excel('products.xlsx')
  purchases = pd.read_excel('/content/purchases.xlsx')
except Error as e:
  print(e)
```

```python
from os import replace
#checking for null
proNullCount= products.isnull().sum()
purNullCount= purchases.isnull().sum()
custNullCount = customers.isnull().sum()

print('Product\n',proNullCount)
print('Purchases\n',purNullCount)
print('customers\n',custNullCount)

print(products[products['company'].isnull()==True])
print('We will make the product with null values as ''ABC''')
products['company']= products['company'].fillna('ABC')
proNullCount= products.isnull().sum()
purNullCount= purchases.isnull().sum()
print('Product\n',proNullCount)
print('Purchases\n',purNullCount)
custNullCount = customers.isnull().sum()
print(custNullCount)
print('for now we will drop the null values from Customers table')
customers.dropna(inplace=True)
print('Shape of customers table: ',customers.shape)
customers.isnull().sum()
purchases['paid'] = purchases['paid'].str.replace('$','')
purchases['paid'] = pd.to_numeric(purchases['paid'], errors='coerce')
```

```python
print('Starting plotting-----------')

gender_counts = customers.gender.value_counts()
plt.pie(gender_counts,labels=gender_counts.index,startangle=120,colors=['red','yellow'],autopct='%1.1f%%')
plt.title('Gender Distribution of Customers')
plt.show()
```

```python
print('City wise customer count')
city_count= customers['city'].value_counts().head(10)
print(city_count)
city_count.plot(kind= 'bar',color='skyblue')
#plt.xlabel('City')
#plt.ylabel('Number of customers')
plt.title('Top cities with mosrt customers')
plt.xticks(rotation= 95)
plt.show()
```

```python
# Create a boxplot of the 'Paid' column
paid = purchases[purchases['paid'] <1000].paid
plt.boxplot(paid)
#plt.ylabel('paid')
plt.title('Boxplot of Amount Paid')
plt.show()
```
