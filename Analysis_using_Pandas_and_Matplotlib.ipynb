```python
import pandas as pd
import matplotlib.pyplot as plt

carData = pd.read_csv('car dataset.csv')
carData.head()
```

Create a bar chart of Count of Cars by Manufacturer. Which manufacturer has the highest number of cars in the dataset?

```python
carManf = carData['Manufacturer'].value_counts().head()
carManf.plot(kind='bar')
plt.ylabel('Quantity of car sold')
plt.show()
print('Hyundai has highest number of cars in the dataset')
```

Create a bar chart to show the average price of cars by category. Which category has the highest average price?

```python
plt.figure(figsize= (6,6))
avg_price_by_category = carData.groupby('Category')['Price'].mean().sort_values(ascending=False)
avg_price_by_category.plot(kind = 'bar',color='green')
plt.ylabel('Average Price')
plt.show()
print('Goods Wagon has the highest average price')
```

Given a dataset in a CSV file named car dataset, create a pie chart to show the top 5 production years with the highest sales. Which of the following years is included in the top 5 years with the highest sales?

```python
topProdYearSales= carData.groupby('Prod. year')['Prod. year'].count().reset_index(name= 'Sales').sort_values(by= 'Sales',ascending=False).head(5)
#can also be found out by using size
topProdYearSales1 = carData.groupby('Prod. year').size().reset_index(name= 'Sales').sort_values(by= 'Sales',ascending=False).head(5)


plt.pie(topProdYearSales1['Sales'],labels= topProdYearSales1['Prod. year'],autopct='%1.1f%%')
plt.title('Top production years in Sales')
plt.axis('equal')
plt.show()
print('Top production years by sales in descending order-',topProdYearSales1['Prod. year'].tolist())
```

Given a dataset in a CSV file named car dataset, create a bar chart to show the count of cars by color. Which color has the most cars?

```python
carColorCount = carData.groupby('Color').size().reset_index(name='count').sort_values(by= 'count',ascending=False).head(5)
print(carColorCount)
plt.bar(carColorCount['Color'],carColorCount['count'],color=carColorCount['Color'],edgecolor='black')
plt.title('Count of car by Color')
plt.show()
print('Black is the colour of most car')
```
