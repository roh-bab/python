```python
import turtle
import time
import random

width,height = 700,600
Colors_list =['red','green','yellow','blue','cyan','violet','pink','orange','brown','purple']
def init_turtle():
    screen = turtle.Screen()
    screen.setup(width,height)
    screen.title('turtle racing')

def number_of_racers():
    racers=0
    while(True):
        racers = input(print("How many turtles you want to race? "))
        print(racers)
        if racers.isdigit():
            racers = int(racers)
        else:
            print("Input is not numeric... Try again!...")
            continue
        if racers<=10 and racers>1:
            return racers
        else:
            print("racers should be 2 to 10")

def createFinishLine():
    print('inside finishLine')
    fin = turtle.Turtle()
    fin.shape('circle')
    fin.color('black')
    fin.speed(1)
    fin.penup()
    fin.setpos(width//2,(height//2)-15)
    fin.pendown()
    fin.backward(710)

def create_turtles(Colors):
    turtles =[]
    spacingx = width//(racers +1)
    createFinishLine()
    for i,color in enumerate(Colors):
        racer = turtle.Turtle()
        racer.color(color)
        racer.shape('turtle')
        racer.penup()
        racer.left(90)
        racer.setpos(-(width+20)//2 + (i + 1) * spacingx, -height//2 + 20)
        turtles.append(racer)
    return turtles

def race(Colours):
    turtles = create_turtles(Colors)
    while True:
        for i,racer in enumerate(turtles):
            distance = random.randrange(1,20)
            racer.pendown()
            racer.forward(distance)
            x, y = racer.pos()
            if y > ((height//2) - 20):
                return i


init_turtle()
racers = number_of_racers()



random.shuffle(Colors_list)
Colors =Colors_list[:racers]
print(Colors)

winner = race(racers)
print("And... the... winner is ",Colors[winner])


```
